## Probability

- $ P(var = value) $: how often does this variable occur with this value
    - Ex. $ P(arm = up) = 0.95 $
- Sometimes notations can be simplified (ex. $ P(up, morning) $)

## Distribution

- How much each value occurs of each variables
    - Ex. night: 0.3, morning: 0.7
    
### Kolmogorov's Aimes

- $ P(A) $ is non-negative real number
- $ P(true) = 1 $ 
    - $ (P(true \land \neg up) = 1 $
- $ P(A \lor B) = P(A) + P(B) - P(A, B) $, A, B mutually exclusive
- Let's say $ x $ has values, p, q, r, $ P(x = p \lor x = q \lor x = r) = 1 $

### Source of Distribution

- Observations
- Scientific beliefs:
    - Causual structures: if iCard is broken, then door does not work
    - Distribution has some standard form: ex. normal distribution
    - Everything is possible $ P(A) \ne 0, P(A) = 0.00000001 $
    - Nothing is guaranteed to be true: $ P(A) \ne 1 $
    
#### Smoothing

Takes observation and shifts probabilities onto what we have not observed
